```python
import numpy as np
import pandas as pd
import pymupdf

import os
print(os.getcwd())
```

```python
doc = pymupdf.open('full_dataset/1-s2.0-S0720048X23000712-main.pdf')
print(doc.metadata)
```

```python
page = doc[0]
text = page.get_text()
print(text)
```

```python
blocks = page.get_text("blocks", sort=True)
for block in blocks:
    print("block #", block[5])
    print(block, '\n') # position, text content, block number, block type
```

```python
pgdict = page.get_text("dict", sort=True) # also includes image blocks

# image: type = 1
# text: type = 0

for block in pgdict['blocks']:
    if(block['type'] == 0):
        print("block #", block['number'])
        for line in block['lines']:
            print("   line dir: ", line['dir'], " wmode: ", line['wmode'], "\n")
            print(line,"\n")
            for span in line['spans']:
                print(" size: ", span['size'], " font: ", span['font'], " color: ", span['color'], "\n")
                print("\t", span['text'], "\n")
```

```python
cropstart = page.search_for("However, they")

for rect in cropstart:
    print(rect)
```

```python
cropend = page.search_for("Fig. 1.")
for rect in cropend:
    print(rect)
```

```python
rx0 = cropstart[0].x0
ry0 = cropstart[0].y0

rx1 = cropstart[0].x1
ry1 = cropstart[0].y1

cr = pymupdf.Rect(rx0,ry0,rx1,ry1)
print(cr)
```

```python
diagram = page.get_text(clip=cr)
print(diagram)
```
