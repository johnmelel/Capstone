```python
import os
from pathlib import Path
from dotenv import load_dotenv
load_dotenv()

doc_dir = "/Users/brunamedeiros/Documents/GitHub/Capstone/Data_Ingestion_VDB/test_dataset"
files = list(Path(doc_dir).glob("*"))
print(files)

file = files[2]
print("\n",file)

from docling.document_converter import DocumentConverter
```

```python
source = file
converter = DocumentConverter()
doc = converter.convert(source).document
```

```python
# Pick which page you want to visualize
page_num = 2   # ðŸ‘ˆ change this

print(f"\n=== TEXT items on page {page_num} ===")
for t in doc.texts:
    if t.prov and t.prov[0].page_no == page_num:
        box = t.prov[0].bbox
        print(f"x0={box.l:.1f}, y0={box.t:.1f}, x1={box.r:.1f}, y1={box.b:.1f}")
        print(t.text[:150].replace("\n", " "))
        print()

print(f"\n=== IMAGE items on page {page_num} ===")
for p in doc.pictures:
    if p.prov and p.prov[0].page_no == page_num:
        box = p.prov[0].bbox
        print(f"x0={box.l:.1f}, y0={box.t:.1f}, x1={box.r:.1f}, y1={box.b:.1f}")
        print("[IMAGE]")
        print()
```
