# Part 2: Images

## 2.1: images

```python
import fitz
from PIL import Image
import io

def extract_images_from_page(page, page_num):
    """Extract all images from a page"""
    images = []
    image_list = page.get_images(full=True)
    
    for img_index, img_info in enumerate(image_list):
        xref = img_info[0]  # Image reference number
        
        # Extract image
        base_image = page.parent.extract_image(xref)
        image_bytes = base_image["image"]
        image_ext = base_image["ext"]  # jpg, png, etc.
        
        # Convert to PIL Image
        image = Image.open(io.BytesIO(image_bytes))
        
        images.append({
            "page": page_num,
            "index": img_index,
            "extension": image_ext,
            "width": base_image["width"],
            "height": base_image["height"],
            "colorspace": base_image["colorspace"],
            "image": image,
            "bytes": image_bytes
        })
    
    return images

# Usage
with fitz.open("/Users/brunamedeiros/Documents/GitHub/Capstone/Fall 2025 Capstone (final folder - edit here)/1_Textbook_Embeddings/full_dataset/1-s2.0-S0720048X23000712-main.pdf") as doc:
    all_images = []
    for page_num, page in enumerate(doc):
        page_images = extract_images_from_page(page, page_num)
        all_images.extend(page_images)
        
    # Save images
    for img_data in all_images:
        filename = f"page_{img_data['page']}_img_{img_data['index']}.{img_data['extension']}"
        img_data['image'].save(filename)
```

## 2.2 images with position

```python
def get_image_positions(page):
    """Get images with their position on the page"""
    image_list = page.get_images(full=True)
    image_positions = []
    
    for img in image_list:
        xref = img[0]
        
        # Get all instances of this image on the page
        rects = page.get_image_rects(xref)
        
        for rect in rects:
            image_positions.append({
                "xref": xref,
                "bbox": (rect.x0, rect.y0, rect.x1, rect.y1),
                "width": rect.width,
                "height": rect.height
            })
    
    return image_positions

# Usage
with fitz.open("/Users/brunamedeiros/Documents/GitHub/Capstone/Fall 2025 Capstone (final folder - edit here)/1_Textbook_Embeddings/full_dataset/1-s2.0-S0720048X23000712-main.pdf") as doc:
    all_images = []
    for page in enumerate(doc):
        page_images = get_image_positions(page)
        all_images.extend(page_images)
        
    # Save images
    for img_data in all_images:
        filename = f"page_{img_data['page']}_img_{img_data['index']}.{img_data['extension']}"
        img_data['image'].save(filename)
```

# 3: Tables

```python
def detect_tables_simple(page):
    """Simple table detection based on text alignment"""
    blocks = page.get_text("blocks")
    
    # Group blocks by y-coordinate (rows)
    rows = {}
    tolerance = 5  # pixels
    
    for block in blocks:
        x0, y0, x1, y1, text, block_no, block_type = block
        if block_type == 0:  # Text block
            # Find or create row
            row_key = None
            for y in rows.keys():
                if abs(y - y0) < tolerance:
                    row_key = y
                    break
            
            if row_key is None:
                row_key = y0
                rows[row_key] = []
            
            rows[row_key].append({
                "x": x0,
                "text": text.strip(),
                "bbox": (x0, y0, x1, y1)
            })
    
    # Sort rows by y-coordinate
    sorted_rows = sorted(rows.items(), key=lambda x: x[0])
    
    # Sort cells in each row by x-coordinate
    table_data = []
    for y, cells in sorted_rows:
        sorted_cells = sorted(cells, key=lambda x: x["x"])
        table_data.append([cell["text"] for cell in sorted_cells])
    
    return table_data
```
